# T1 - Shape subdivision
This tutorial shows how to use split shapes into Voronoi sub-shapes

```python
import time
import rastertools
from rastertools import shape_subdivide
from rastertools.shape import plot_shapes
from pathlib import Path
print(f"rastertools version v{rastertools.__version__}")
```

RasterTools includes a function to subdivide a shept based on a target area. Here is how you call it:

```python
def subdivide_example(area: int = None, shape_file=Path('data/COD_LEV02_ZONES')):
    start_time = time.time()

    print(f"Starting {area or 'default'} subdivision...")
    new_shape_stem = shape_subdivide(shape_stem=shape_file, out_dir="results",
                                     box_target_area_km2=area, verbose=False)
    print(f"Completed subdivision in {round(time.time() - start_time)}s")

    return shape_file, new_shape_stem
```

Running this with the default (100 square km):

```python
shape_file, new_shape_stem = subdivide_example(900)  # default is 100 km2
```

RasterTools also includes a function you can use to plot shapefiles:

```python
fig, ax = plot_shapes(shape_file, alpha=1.0, edgecolor='k', facecolor='gray')
plot_shapes(new_shape_stem, ax=ax, alpha=1.0, linewidth=0.2, edgecolor='k', facecolor='None')
ax.set_box_aspect(1)
```
