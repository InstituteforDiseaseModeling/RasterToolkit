```python
from pathlib import Path
from rastertools import download, raster_clip, utils
import numpy as np
import rasterio
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
population_raster_file = Path('datasets\\pop_0to40.tif')
shape_file = Path('datasets\\nga_admbnda_adm1_osgof\\nga_admbnda_adm1_osgof_20161215')
```

```python
pop_dict2 = raster_clip(population_raster_file, shape_file, include_latlon=True, shape_attr='admin1Name')
```

```python
df = pd.DataFrame.from_dict(pop_dict2, orient="index")
df = df.reset_index(names=["admin1Name"])
df.to_csv("results/clipped_pop_nigeria_admin1.csv", index=False)
```
